```python
import pandas as pd
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv('data/df_linreg.csv')
df = df.rename(columns={"0": "qtd"})

df.head()
```

```python
model = ols("qtd ~ pais_origem + pais_destino + year + month + day",  data=df)
response = model.fit()
response.summary()
```

```python
model = ols("qtd ~ pais_origem + pais_destino + year + month + day",  data=df[:10000])
response = model.fit()
response.summary()
```
